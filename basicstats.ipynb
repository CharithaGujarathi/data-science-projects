```python
import seaborn as sns
import matplotlib.pyplot as plt


categorical_cols = df.select_dtypes(include=['object', 'category']).columns


for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df, palette='Set2')
    plt.title(f'Bar Chart for {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()


```

```python
Some items or regions show up a lot more often than others in the chart. This means a few products or areas are doing most of the business, so it makes sense to focus more on those to grow sales.
```

# STANDARTIZATION OF NUMERICAL VARIABLES

```python
pip install seaborn
```

EXPLANATION:
Standardization is a way to make different types of numbers easier to compare. 
For example, prices might be in thousands and discounts in single diGITS, that means they are different scales. So, we use a formula called z-score to scale them.

in standertization we subtract the mean from each value and then divide by standard deviation. After standardization, all numbers come to the same scale they’re centered around 0 and usually fall between -3 and +3.

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = df.select_dtypes(include='number').columns

scaler = StandardScaler()

df2 = df.copy()  
df2[numeric_cols] = scaler.fit_transform(df2[numeric_cols])

print(df2.head())
```

#Conversion of Categorical Data into Dummy Variables

Models can't understand text, so we turn categories into numbers using dummy variables. Each category becomes a column with 0s and 1s, helping the model read and learn from the data.

```python
import pandas as pd

categorical_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns

df3 = pd.get_dummies(df, columns=categorical_cols, drop_first=False)

df3 = df3.astype(int)

print(df3.head())
```
